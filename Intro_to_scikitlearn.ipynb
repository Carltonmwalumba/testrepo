## 0. An end-to-end sciKit-Learn workflow

```python
# 1. Get the data ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# Create X ie features matrix
X = heart_disease.drop("target", axis = 1)

# create Y(labels)
y = heart_disease["target"]
```

```python
#2. Choose the right Models and type of hyperparameters
#We've Chosen a Classification Model ie, RandomForest
from sklearn.ensemble import RandomForestClassifier # Classification Ml Model
clf = RandomForestClassifier()

#Well keep the deafult hyperparameters
clf.get_params()
```

```python
#3. Fit the model to the training data
from sklearn.model_selection import train_test_split

#Split the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf.fit(X_train, y_train)
```

```python
# Make a prediction
# Note that the shape of the data to be predicted should match the training data shape
y_label = clf.predict(np.array([0, 2, 3, 4]))
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the Model on the training data and test data
#On training data
clf.score(X_train, y_train)
```

```python
#On test data
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve the model
# Try differnt amounts of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf, open("rando_forest_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("rando_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```

```python
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales.dtypes
```

```python
#Doors is categorical as different models can fit into different categories
car_sales["Doors"].value_counts()
```

```python
# Split to X/y
X = car_sales.drop("Price", axis = 1)
y = car_sales["Price"]

# Split into training and test
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2)
```

```python
#Build a machine learning model
from sklearn.ensemble import RandomForestRegressor

#Define model
model = RandomForestRegressor()
#Fit model
model.fit(X_train, y_train)
#Evaluate model
model.score(X_test, y_test)
```

```python
#Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Define our Categorical Features
categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                               remainder = "passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
```
